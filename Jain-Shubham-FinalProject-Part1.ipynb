```python
#reading the CSV indivvidually and creating a super dataset
data_2018 = pd.read_csv('https://raw.githubusercontent.com/ShubhamJain2798/IS445-DataViz-IndiaExportData/main/2018.csv')
data_2018['YEAR'] = 2018
data_2018.rename(columns={"UNIT_CODE": "UNIT", "QTY": "QUANTITY"})
data_2018

#here i tried to rename the columns but it did not work
```

```python
#trying different method to rename but still did not work
data_2018.rename({"UNIT_CODE":"UNIT", "QTY":"QUANTITY"}, axis='columns')
data_2018
```

```python
#method which worked
data_2018.rename(columns={"UNIT_CODE":"UNIT", "QTY":"QUANTITY","VALUES (US$ million)":"VALUE (US$ million)"}, inplace=True)
data_2018
```

```python
column_list = list(data_2018.columns)
column_list
```

```python
data_2019 = pd.read_csv('https://raw.githubusercontent.com/ShubhamJain2798/IS445-DataViz-IndiaExportData/main/2019.csv')
data_2019['YEAR'] = 2019
data_2019 = data_2019.reindex(column_list, axis='columns')
data_2019
```

```python
data_2020 = pd.read_csv('https://raw.githubusercontent.com/ShubhamJain2798/IS445-DataViz-IndiaExportData/main/2020.csv')
data_2020['YEAR'] = 2020
data_2020 = data_2020.reindex(column_list, axis='columns')
data_2020
```

```python
#adding data of all three tables to one table

df = data_2018
df = df.append(data_2019, ignore_index=True).append(data_2020, ignore_index=True)
df.value_counts()
```

```python
data_grpby_commodity = df.loc[:, df.columns!='YEAR'].groupby(['COMMODITY']).sum().sort_values('VALUE (US$ million)',ascending = False).head(10)
data_grpby_commodity.reset_index(level=0, inplace=True)
data_grpby_commodity
```

```python
with plt.style.context("bmh"):
    fig, ax = plt.subplots(2, figsize=(20,20))
    
    X1 = data_grpby_commodity['COMMODITY']
    Y1 = data_grpby_commodity['QUANTITY']
    
    X2 = data_grpby_commodity['COMMODITY']
    Y2 = data_grpby_commodity['VALUE (US$ million)']
    
    ax[0].tick_params(labelrotation=90)
    ax[1].tick_params(labelrotation=90)

    
    ax[0].bar(X1, Y1)
    ax[1].bar(X2, Y2)
```

```python
#Total row data does not make sense at all so removing it:
data_grpby_commodity.drop(data_grpby_commodity[data_grpby_commodity.COMMODITY == 'Total'].index, inplace=True)
with plt.style.context("bmh"):
    fig, ax = plt.subplots(2, figsize=(20,20))
    
    X1 = data_grpby_commodity['COMMODITY']
    Y1 = data_grpby_commodity['QUANTITY']
    
    X2 = data_grpby_commodity['COMMODITY']
    Y2 = data_grpby_commodity['VALUE (US$ million)']
    
    ax[0].tick_params(labelrotation=90)
    ax[1].tick_params(labelrotation=90)

    ax[0].set(xlabel="COMMODITY", ylabel="QUANTITY EXPORTED")
    ax[1].set(xlabel="COMMODITY", ylabel="VALUE (US$ million)")
    
    ax[0].set_title('Cumulative quantity for top 10 valuable commodities 2018-2020')
    ax[1].set_title('Cumulative export value for top 10 valuable commodities 2018-2020')
    
    ax[0].bar(X1, Y1)
    ax[1].bar(X2, Y2)
    
    fig.tight_layout(pad=10.0)

    
#Here we get to know that even if the quantity for petroleum products is the least, it is the most valueable export commodity

```

```python
data_grpby_commodity_year = df.groupby(['COMMODITY','YEAR','UNIT']).sum()
data_grpby_commodity_year.reset_index(level=[0,1,2], inplace=True)
data_grpby_commodity_year.drop(data_grpby_commodity_year[data_grpby_commodity_year.COMMODITY == 'Total'].index, inplace=True)
# data_grpby_commodity_year.reset_index(level=1, inplace=True)
data_grpby_commodity_year
```

```python
#but I realized this does not make sense as the quantity has different measuring units like KGS etc


commodities = list(data_grpby_commodity_year['COMMODITY'].unique())

@widgets.interact(product = commodities)
def update(product='PETROLEUM PRODUCTS'):
    t_data = data_grpby_commodity_year[data_grpby_commodity_year['COMMODITY'] == product]
    fig, ax = plt.subplots(1,2, figsize=(16,9))
    
    unit = t_data['UNIT'].unique()[0]
    years = list(t_data['YEAR'].unique())
    
    X1 = t_data['YEAR']
    Y1 = t_data['QUANTITY']
    
    X2 = t_data['YEAR']
    Y2 = t_data['VALUE (US$ million)']
    
    
    ax[0].set_xticks(years)
    ax[1].set_xticks(years)

    ax[0].tick_params(labelrotation=90)
    ax[1].tick_params(labelrotation=90)

    ax[0].set(xlabel="COMMODITY", ylabel="QUANTITY EXPORTED," + " UNIT: "+unit)
    ax[1].set(xlabel="COMMODITY", ylabel="VALUE (US$ million)")
    
    
    ax[0].set_title('Yearly quantity for '+ product)
    ax[1].set_title('Yearly export value for '+ product)
    
    ax[0].bar(X1, Y1)
    ax[1].bar(X2, Y2)
    
    fig.tight_layout(pad=10.0)
#     return fig
```
